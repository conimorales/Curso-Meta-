Este curso es el primero de una serie que pretende ayudarle a aprender más sobre el desarrollo web y le prepara para utilizar Bootstrap en una página biográfica que creará. Al final de este curso, será capaz de:

  - Describir el papel del desarrollador front-end

  - Explicar las tecnologías básicas y subyacentes que impulsan Internet

  - Utilizar HTML para crear una página web sencilla

  - Utilizar CSS para controlar la apariencia de una página web sencilla

  - Explicar qué es React

  - Describir las aplicaciones y características de los frameworks de interfaz de usuario más populares

En este curso, explorará lo siguiente:

Módulo 1: Introducción al desarrollo web
En este módulo, se le introducirá en el desarrollo web. Conocerá los distintos tipos de funciones de los desarrolladores web y las responsabilidades de los desarrolladores front-end, back-end y full-stack. Obtendrá una visión general de las tecnologías básicas de HTML, CSS y JavaScript y explorará los conceptos que sustentan el funcionamiento de Internet. Además, podrá acceder a ejercicios prácticos para editar un sitio web. 

Tras completar este módulo, será capaz de: 

Describir el papel laboral del desarrollador web. 

Distinguir entre desarrolladores front-end, back-end y full-stack.

Explicar cómo se mueven los datos por Internet.

Describir las tecnologías que sustentan Internet.

Módulo 2: Introducción a HTML5 y CSS

Aquí aprenderá sobre HTML5 y CSS. También examinará cómo construir documentos HTML y añadir estilos y diseños básicos utilizando CSS. 

Tras completar este módulo, será capaz de: 

Utilizar HTML para crear una página web sencilla.

Utilizar CSS para definir el estilo de una página web sencilla.

Módulo 3: Marcos de interfaz de usuario
En este módulo, aprenderá sobre los frameworks de interfaz de usuario. Además, aprenderá a utilizar el framework Bootstrap para construir interfaces responsivas. Explorará las ventajas de trabajar con marcos de interfaz de usuario. 

Tras completar este módulo, será capaz de: 

Esbozar los conceptos que existen en la mayoría de los marcos de interfaz de usuario.

Utilizar el marco CSS de Bootstrap para crear páginas web.

Aprovechar la documentación de Bootstrap para reproducir y modificar componentes CSS.

Utilizar temas de Bootstrap. 

Describir los fundamentos de React en relación con otros frameworks y tecnologías web.

 Módulo 4: Evaluación graduada
Aquí aprenderá sobre la evaluación graduada. Después de completar las unidades individuales de este módulo, sintetizará las habilidades del curso para crear y dar estilo a una página biográfica. También tendrá la oportunidad de reflexionar sobre el contenido del curso y el camino de aprendizaje que tiene por delante. 

Tras completar este módulo, será capaz de: 

Crear y dar estilo a una página biográfica.

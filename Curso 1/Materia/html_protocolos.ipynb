PROTOCOLOS 

Los Protocolos de transferencia de hipertexto (HTTP) se utilizan sobre el Protocolo de control de transmisin (TCP) para transferir pginas web y otros contenidos desde sitios web.
Esta lectura explora otros protocolos utilizados habitualmente en Internet.

Protocolo de configuracin dinmica de host (DHCP)
Ya has aprendido que los ordenadores necesitan direcciones IP para comunicarse entre s. Cuando su ordenador se conecta a una red, el Protocolo de configuracin dinmica de host o DHCP, como se conoce comnmente, se utiliza para asignar a su ordenador una direccin IP.
Su ordenador se comunica a travs del Protocolo de Datagramas de Usuario (UDP) utilizando el protocolo con un tipo de servidor llamado servidor DHCP. El servidor lleva un registro de los ordenadores de la red y sus direcciones IP. Asignar a su ordenador una direccin IP y responder a travs del protocolo para indicarle qu direccin IP debe utilizar. Una vez que tu ordenador tiene una direccin IP, puede comunicarse con otros ordenadores de la red.

Protocolo del sistema de nombres de dominio (DNS)
Su ordenador necesita una forma de saber con qu direccin IP debe comunicarse cuando visita un sitio web en su navegador, por ejemplo, meta.com. El Protocolo del Sistema de Nombres de Dominio, comnmente conocido como DNS, proporciona esta funcin. A continuacin, su ordenador consulta el servidor DNS asociado al nombre de dominio y le devuelve la direccin IP correcta.

Protocolo de Acceso a Mensajes de Internet (IMAP)
Compruebas tu correo electrnico en tu mvil o tableta? O tal vez utilizas una aplicacin de correo electrnico en tu ordenador?
Su dispositivo necesita una forma de descargar correos electrnicos y gestionar su buzn en el servidor que almacena sus correos electrnicos. Este es el propsito del Protocolo de Acceso a Mensajes de Internet o IMAP.

Protocolo simple de transmisin de correo (SMTP)
Ahora que tus correos electrnicos estn en tu dispositivo, necesitas una forma de enviarlos. Para ello se utiliza el Protocolo simple de transmisin de correo o SMTP. Permite a los clientes de correo electrnico enviar correos electrnicos a travs de un servidor SMTP. Tambin puede utilizarlo para recibir correos electrnicos de un cliente de correo electrnico, pero IMAP es ms comnmente utilizado.

Protocolo de oficina de correos (POP)
El Protocolo de Oficina de Correos (POP) es un protocolo ms antiguo utilizado para descargar correos electrnicos a un cliente de correo electrnico. La principal diferencia entre POP e IMAP es que POP elimina los mensajes del servidor una vez descargados en el dispositivo local. Aunque ya no se utiliza habitualmente en los clientes de correo electrnico, los desarrolladores suelen utilizarlo para implementar la automatizacin del correo electrnico, ya que es un protocolo ms sencillo que IMAP.

Protocolo de transferencia de archivos (FTP)
Cuando ejecutes tus sitios y aplicaciones web en Internet, necesitars una forma de transferir los archivos desde tu ordenador local al servidor en el que se ejecutarn. El protocolo estndar utilizado para ello es el Protocolo de transferencia de archivos o FTP. FTP permite listar, enviar, recibir y borrar archivos en un servidor. Tu servidor debe ejecutar un Servidor FTP y necesitars un Cliente FTP en tu mquina local. Aprenders ms sobre esto en un curso posterior.

Protocolo Secure Shell (SSH)
Cuando empieces a trabajar con servidores, tambin necesitars una forma de iniciar sesin e interactuar con el ordenador de forma remota. El mtodo ms comn de hacer esto es usando el Protocolo Secure Shell, comnmente referido como SSH. El uso de un cliente SSH le permite conectarse a un servidor SSH que se ejecuta en un servidor para ejecutar comandos en el equipo remoto.
Todos los datos enviados a travs de SSH estn encriptados. Esto significa que terceros no pueden entender los datos transmitidos. Slo los ordenadores emisor y receptor pueden entender los datos.

Protocolo seguro de transferencia de archivos (SFTP)
Los datos se transmiten de forma insegura cuando se utiliza el Protocolo de transferencia de archivos. Esto significa que terceros pueden entender los datos que usted est enviando. Esto no es RIGHT si transmites archivos de empresa como software y bases de datos. Para solucionar esto, se puede utilizar el Protocolo seguro de transferencia de archivos SSH, tambin llamado Protocolo seguro de transferencia de archivos, para transferir archivos a travs del protocolo SSH. Esto garantiza que los datos se transmiten de forma segura. La mayora de los clientes FTP tambin son compatibles con el protocolo SFTP.
